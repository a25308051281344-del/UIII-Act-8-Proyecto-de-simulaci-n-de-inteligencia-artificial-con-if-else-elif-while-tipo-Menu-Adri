```python
# ==========================================
# IA COSTURERA Y MÁQUINA DE COSER
# Nivel: Principiante
# Utiliza: if, else, elif y while True
# No usa clases, funciones, for ni try
# ==========================================

print("====================================")
print(" IA COSTURERA Y MÁQUINA DE COSER ")
print("====================================")

while True:

    print("\n----------- MENÚ -----------")
    print("1. Confeccionar Camisa")
    print("2. Confeccionar Pantalón")
    print("3. Confeccionar Vestido")
    print("4. Reparar Ropa")
    print("5. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":

        print("\n--- CAMISA ---")

        tela = input("¿La tela es algodón? (si/no): ")
        urgente = input("¿Es urgente? (si/no): ")

        if tela == "si" and urgente == "si":
            print("IA: Utilizar máquina rápida.")
            print("IA: Aplicar costura reforzada.")

        elif tela == "si":
            print("IA: Utilizar máquina estándar para algodón.")

        else:
            print("IA: Revisar el tipo de tela antes de coser.")

    elif opcion == "2":

        print("\n--- PANTALÓN ---")

        bolsas = input("¿Lleva bolsas? (si/no): ")
        mezclilla = input("¿Es mezclilla? (si/no): ")

        if bolsas == "si" and mezclilla == "si":
            print("IA: Utilizar aguja resistente.")
            print("IA: Agregar doble costura.")

        elif mezclilla == "si":
            print("IA: Utilizar máquina para tela gruesa.")

        else:
            print("IA: Costura normal recomendada.")

    elif opcion == "3":

        print("\n--- VESTIDO ---")

        elegante = input("¿Es para una fiesta? (si/no): ")
        encaje = input("¿Lleva encaje? (si/no): ")

        if elegante == "si" and encaje == "si":
            print("IA: Recomienda acabados finos.")
            print("IA: Costura decorativa especial.")

        elif elegante == "si":
            print("IA: Utilizar diseño elegante.")

        else:
            print("IA: Diseño casual recomendado.")

    elif opcion == "4":

        print("\n--- REPARACIÓN DE ROPA ---")

        dano = input("¿La prenda está muy dañada? (si/no): ")

        if dano == "si":
            print("IA: Reparación completa recomendada.")

        else:
            print("IA: Reparación sencilla recomendada.")

    elif opcion == "5":

        print("\nGracias por utilizar la IA Costurera.")
        print("Programa finalizado.")
        break

    else:

        print("\nOpción incorrecta.")
        print("Seleccione una opción válida.")
```
